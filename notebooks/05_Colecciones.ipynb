# 5) Colecciones: listas, tuplas y diccionarios

## ¿Qué son?
Estructuras para **agrupar datos**: secuencias (`list`, `tuple`) y mapeos (`dict`).

## ¿Para qué se usan?
Para **organizar, acceder y transformar** conjuntos de valores.

## Puntos clave
- `list` (mutable) — métodos como `append`, `pop`, `sort`.
- `tuple` (inmutable) — útil para datos que no cambian.
- `dict` (clave→valor) — acceso por clave y métodos `.items()`, `.keys()`.


```python
comidas = ["arepa", "ajiaco", "bandeja"]
comidas.append("sancocho")
print("Lista:", comidas)

punto = (10, 20)
print("Tupla:", punto[0], punto[1])

persona = {"nombre": "Ada", "edad": 28, "ciudad": "Bogotá"}
persona["edad"] += 1
for k, v in persona.items():
    print(k, "->", v)
```

### Reto extra
**Inventario simple**: lista de dicts con `{"producto","precio","stock"}`. Muestra:
- Total del inventario.
- Producto más caro.


```python
# Ejercicios sobre Colecciones

# E1: Lista de 5 números, calcula suma y promedio.
print("--- E1: Suma y Promedio de una Lista ---")
numeros = [15, 42, 8, 99, 23]
suma = sum(numeros)
promedio = suma / len(numeros)

print(f"La lista es: {numeros}")
print(f"La suma es: {suma}")
print(f"El promedio es: {promedio}")
print("-" * 20)


# E2: Diccionario de 3 contactos (nombre→tel), consulta por nombre.
print("--- E2: Agenda de Contactos ---")
contactos = {
    "Ana": "555-1234",
    "Luis": "555-5678",
    "Eva": "555-8765"
}

nombre_a_buscar = input("Introduce el nombre del contacto a buscar: ")
telefono = contactos.get(nombre_a_buscar, "Contacto no encontrado.")

print(f"El teléfono de {nombre_a_buscar} es: {telefono}")
print("-" * 20)


# E3: Convierte lista de pares (clave, valor) en dict y recórrelo.
print("--- E3: Convertir Pares a Diccionario ---")
lista_de_pares = [
    ("producto", "Laptop"),
    ("precio", 1200),
    ("disponible", True)
]

# Convertir la lista de tuplas a un diccionario
diccionario_producto = dict(lista_de_pares)

print("Diccionario creado:")
for clave, valor in diccionario_producto.items():
    print(f"- {clave}: {valor}")
print("-" * 20)


```

```python
# Reto extra: Inventario simple
print("--- Reto Extra: Inventario Simple ---")
inventario = [
    {"producto": "Manzanas", "precio": 2.5, "stock": 100},
    {"producto": "Naranjas", "precio": 3.0, "stock": 80},
    {"producto": "Plátanos", "precio": 1.8, "stock": 120}
]

# Calcular el valor total del inventario
valor_total = 0
for item in inventario:
    valor_total += item["precio"] * item["stock"]

# Encontrar el producto más caro
producto_mas_caro = None
precio_maximo = -1 # Iniciar con un valor bajo

for item in inventario:
    if item["precio"] > precio_maximo:
        precio_maximo = item["precio"]
        producto_mas_caro = item

print(f"El valor total del inventario es: ${valor_total:.2f}")

if producto_mas_caro:
    print(f"El producto más caro es '{producto_mas_caro['producto']}' con un precio de ${producto_mas_caro['precio']:.2f}")
else:
    print("El inventario está vacío.")
print("-" * 20)
```
