# 3) Operadores y expresiones

## ¿Qué son?
Los **operadores** combinan valores para producir un resultado (*expresiones*).

## ¿Para qué se usan?
Para **calcular, comparar y tomar decisiones**. Son la base de toda lógica.

## Familias principales
- Aritméticos: + - * / // % **
- Comparación: == != > < >= <=
- Lógicos: and or not
- Precedencia: paréntesis > potencias > multiplicación/división > suma/resta


```python
a, b = 10, 3
print("Aritméticos:", a + b, a - b, a * b, a / b, a // b, a % b, a ** b)
print("Comparación:", a == b, a != b, a > b, a <= b)
x, y = True, False
print("Lógicos:", x and y, x or y, not x)
```

### Reto extra
Calculadora por consola que reciba a, b y una operación, y muestre el resultado.


```python
# Pista: calculadora básica
a, b, op = 12, 5, "*"
if op == "+":
    print(a + b)
elif op == "-":
    print(a - b)
elif op == "*":
    print(a * b)
elif op == "/":
    print(a / b)
elif op == "//":
    print(a // b)
elif op == "%":
    print(a % b)
elif op == "**":
    print(a ** b)
else:
    print("Operación no reconocida")
```

```python
# Ejercicios sobre Operadores y Expresiones

# E1: Área de un círculo con pi=3.14159 y radio fijo.
pi = 3.14159
radio = 7.5
area = pi * (radio ** 2)
print(f"El área de un círculo con radio {radio} es: {area:.2f}")
print("-" * 20)

# E2: Verifica si un número es múltiplo de 3 y 5.
numero = 30
es_multiplo = (numero % 3 == 0) and (numero % 5 == 0)
print(f"¿Es {numero} múltiplo de 3 y 5? {es_multiplo}")

numero = 10
es_multiplo = (numero % 3 == 0) and (numero % 5 == 0)
print(f"¿Es {numero} múltiplo de 3 y 5? {es_multiplo}")
print("-" * 20)

# E3: Evalúa si una edad está en el rango 18–65, pidiendo la edad al usuario.

try:
    # Pide la edad y la convierte a entero
    edad_str = input("Introduce tu edad: ")
    edad = int(edad_str)
    
    # Evalúa si la edad está en el rango
    esta_en_rango = 18 <= edad <= 65
    
    # Muestra un mensaje personalizado según el resultado
    if esta_en_rango:
        print(f"La edad {edad} está dentro del rango laboral (18-65).")
    else:
        print(f"La edad {edad} está fuera del rango laboral (18-65).")

except ValueError:
    # Se ejecuta si el usuario no introduce un número
    print("Entrada no válida. Por favor, introduce un número entero para la edad.")
```

```python
# Reto extra: Calculadora por consola

try:
    # Pedir los números y el operador al usuario
    a_str = input("Introduce el primer número (a): ")
    b_str = input("Introduce el segundo número (b): ")
    op = input("Introduce la operación (+, -, *, /, //, %, **): ")

    # Convertir los números a flotantes para aceptar decimales
    a = float(a_str)
    b = float(b_str)

    resultado = None
    # Realizar la operación correspondiente
    if op == "+":
        resultado = a + b
    elif op == "-":
        resultado = a - b
    elif op == "*":
        resultado = a * b
    elif op == "/":
        # Manejar la división por cero
        if b == 0:
            print("Error: No se puede dividir por cero.")
        else:
            resultado = a / b
    elif op == "//":
        # Manejar la división por cero
        if b == 0:
            print("Error: No se puede dividir por cero.")
        else:
            resultado = a // b
    elif op == "%":
        # Manejar la división por cero
        if b == 0:
            print("Error: No se puede dividir por cero.")
        else:
            resultado = a % b
    elif op == "**":
        resultado = a ** b
    else:
        print("Operación no reconocida.")

    # Imprimir el resultado solo si la operación fue exitosa
    if resultado is not None:
        print(f"El resultado de {a} {op} {b} es: {resultado}")

except ValueError:
    # Se ejecuta si la conversión a float falla
    print("Error: Asegúrate de introducir solo números para 'a' y 'b'.")
```
