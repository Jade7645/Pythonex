# 2) Tipos de datos y variables

## ¿Qué son?
Los **tipos** definen la clase de datos (entero, flotante, texto, booleano). Las **variables** son nombres que apuntan a valores en memoria.

## ¿Para qué se usan?
Para **representar y transformar información**. Python es *tipado dinámico*: el tipo se asocia al valor, no al nombre.

## Tipos básicos
- int, float, str, bool
- Mutabilidad: str es inmutable, list es mutable.
- Conversión de tipos: int("42"), float("3.14"), str(99).


```python
edad = 28
altura = 1.72
nombre = "Ada"
es_estudiante = True

print(type(edad), type(altura), type(nombre), type(es_estudiante))
print(f"{nombre} mide {altura} m y tiene {edad} años. ¿Estudia? {es_estudiante}")
```

### Reto extra
Calcula el **IMC**: pide peso (kg) y altura (m), imprime el resultado con dos decimales.


```python
# Plantilla para IMC
peso = 70.0
altura = 1.75
imc = peso / (altura ** 2)
print(f"IMC: {imc:.2f}")
```

```python
# Ejercicios sobre Tipos y Variables

# E1: Declara tres variables de distintos tipos y muéstralas con su tipo.
variable_texto = "Python"
variable_entero = 2024
variable_flotante = 3.1416

print(f"'{variable_texto}' es de tipo {type(variable_texto)}")
print(f"{variable_entero} es de tipo {type(variable_entero)}")
print(f"{variable_flotante} es de tipo {type(variable_flotante)}")
print("-" * 20)

# E2: Convierte "99" a entero y súmalo con 1.
numero_como_cadena = "99"
resultado = int(numero_como_cadena) + 1
print(f'El resultado de convertir "99" a entero y sumarle 1 es: {resultado}')
print("-" * 20)

# E3: Pide por input() nombre y edad, y muestra un mensaje formateado.
input_nombre = input("Por favor, introduce tu nombre: ")
input_edad = input("Ahora, introduce tu edad: ")
print(f"¡Hola, {input_nombre}! Veo que tienes {input_edad} años.")
```

```python
# Reto extra: Calculadora de IMC (versión simplificada)

# Pedir peso y altura al usuario
peso_str = input("Introduce tu peso en kg: ")
altura_str = input("Introduce tu altura en metros: ")

# Convertir las entradas a números flotantes
peso = float(peso_str)
altura = float(altura_str)

# Calcular el IMC
imc = peso / (altura ** 2)

# Imprimir el resultado formateado con dos decimales
print(f"Tu Índice de Masa Corporal (IMC) es: {imc:.2f}")
```
