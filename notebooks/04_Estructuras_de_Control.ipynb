# 4) Estructuras de control: if, for, while

## ¿Qué son?
Mecanismos para **tomar decisiones** (if) y **repetir acciones** (for, while).

## ¿Para qué se usan?
Para **controlar el flujo** de un programa según condiciones y colecciones.

## Ideas clave
- La **indentación** define bloques.
- for itera sobre secuencias; range(n) genera 0..n-1.
- while repite mientras la condición sea verdadera.
- break/continue controlan el bucle.


```python
# Par o impar
n = 7
if n % 2 == 0:
    print("Par")
else:
    print("Impar")

# Suma 1..10
total = 0
for i in range(1, 11):
    total += i
print("Suma 1..10:", total)

# Cuenta regresiva con while
c = 5
while c > 0:
    print(c)
    c -= 1
```

### Ejercicios
- E1: Clasifica una nota (0–100) en A/B/C/D/F.
- E2: Imprime la tabla de multiplicar de un número (1–10).
- E3: Con while, pide números hasta 0; muestra suma y promedio.

### Reto extra
**FizzBuzz** de 1 a 100.


```python
# Ejercicios sobre Estructuras de Control (versión simplificada)

# E1: Clasifica una nota (0–100) en A/B/C/D/F.
print("--- E1: Clasificador de Notas ---")
nota_str = input("Introduce una nota del 0 al 100: ")
nota = int(nota_str)

if nota > 100 or nota < 0:
    print("La nota debe estar entre 0 y 100.")
elif nota >= 90:
    print(f"Una nota de {nota} es una 'A'.")
elif nota >= 80:
    print(f"Una nota de {nota} es una 'B'.")
elif nota >= 70:
    print(f"Una nota de {nota} es una 'C'.")
elif nota >= 60:
    print(f"Una nota de {nota} es una 'D'.")
else:
    print(f"Una nota de {nota} es una 'F'.")
print("-" * 20)


# E2: Imprime la tabla de multiplicar de un número (1–10).
print("--- E2: Tabla de Multiplicar ---")
num_str = input("Introduce un número para ver su tabla de multiplicar: ")
numero = int(num_str)

print(f"Tabla del {numero}:")
for i in range(1, 11):
    print(f"{numero} x {i} = {numero * i}")
print("-" * 20)


# E3: Con while, pide números hasta 0; muestra suma y promedio.
print("--- E3: Suma y Promedio (termina con 0) ---")
numeros = []
while True:
    num_input_str = input("Introduce un número (o 0 para terminar): ")
    num_input = int(num_input_str)
    
    if num_input == 0:
        break
    numeros.append(num_input)

if numeros:
    suma_total = sum(numeros)
    promedio = suma_total / len(numeros)
    print(f"Números introducidos: {numeros}")
    print(f"Suma total: {suma_total}")
    print(f"Promedio: {promedio:.2f}")
else:
    print("No se introdujeron números para calcular.")
print("-" * 20)


# Reto extra: FizzBuzz de 1 a 100.
print("--- Reto Extra: FizzBuzz ---")
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz", end=", ")
    elif i % 3 == 0:
        print("Fizz", end=", ")
    elif i % 5 == 0:
        print("Buzz", end=", ")
    else:
        print(i, end=", ")
print("\n" + "-" * 20)
```

```python
# Reto extra: FizzBuzz de 1 a 100

print("--- Reto Extra: FizzBuzz ---")

# Itera a través de los números del 1 al 100
for i in range(1, 101):
    # Si es múltiplo de 3 y 5 (o sea, de 15)
    if i % 15 == 0:
        print("FizzBuzz", end=", ")
    # Si es solo múltiplo de 3
    elif i % 3 == 0:
        print("Fizz", end=", ")
    # Si es solo múltiplo de 5
    elif i % 5 == 0:
        print("Buzz", end=", ")
    # Si no es múltiplo de ninguno
    else:
        print(i, end=", ")

# Imprime un salto de línea al final para un formato limpio
print("\n" + "-" * 20)
```
