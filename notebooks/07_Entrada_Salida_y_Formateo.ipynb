# 7) Entrada/Salida y formateo

## ¿Qué es?
La **interacción por consola**: leer con input() y mostrar con print().

## ¿Para qué se usa?
Para **recoger datos del usuario** y **presentar resultados** con buen formato.

## Claves de formateo
- f-strings: f"{valor:.2f}"
- sep y end en print
- Conversión explícita de tipos al leer (int, float)


```python
# Suma con formateo
a = 12.3456
b = 7.8
print(f"Suma: {a + b:.2f}")
print("uno", "dos", "tres", sep=" | ", end=".
")
```

### Ejercicios
- E1: Pide nombre y año de nacimiento; calcula edad.
- E2: Factura: subtotal, IVA (19%), total con 2 decimales.
- E3: “Ficha” multi-línea con \n y f-strings.

### Reto extra
Lee tres notas y muestra promedio, mínima, máxima y si aprueba (>=3.0).


```python
# Ejercicios sobre Entrada/Salida y Formateo

# E1: Pide nombre y año de nacimiento; calcula edad.
print("--- E1: Calculadora de Edad ---")
from datetime import datetime

nombre = input("Introduce tu nombre: ")
try:
    año_nacimiento_str = input("Introduce tu año de nacimiento: ")
    año_nacimiento = int(año_nacimiento_str)
    
    año_actual = datetime.now().year
    edad = año_actual - año_nacimiento
    
    print(f"Hola, {nombre}. Tienes o cumplirás {edad} años en {año_actual}.")
except ValueError:
    print("El año de nacimiento debe ser un número válido.")
print("-" * 20)


# E2: Factura: subtotal, IVA (19%), total con 2 decimales.
print("--- E2: Cálculo de Factura ---")
try:
    subtotal_str = input("Introduce el subtotal de la factura: ")
    subtotal = float(subtotal_str)
    
    IVA_TASA = 0.19
    iva_monto = subtotal * IVA_TASA
    total = subtotal + iva_monto
    
    print("\n--- Detalle de la Factura ---")
    print(f"Subtotal: ${subtotal:10.2f}")
    print(f"IVA (19%): ${iva_monto:10.2f}")
    print("-------------------------")
    print(f"Total:    ${total:10.2f}")
except ValueError:
    print("El subtotal debe ser un valor numérico.")
print("-" * 20)


# E3: “Ficha” multi-línea con \n y f-strings.
print("--- E3: Ficha de Presentación ---")
ficha_nombre = "Ada Lovelace"
ficha_profesion = "Matemática y Programadora"
ficha_ciudad = "Londres"

ficha_completa = (
    f"--- Ficha de Usuario ---\n"
    f"Nombre:    {ficha_nombre}\n"
    f"Profesión: {ficha_profesion}\n"
    f"Ciudad:    {ficha_ciudad}\n"
    f"------------------------"
)
print(ficha_completa)
print("-" * 20)


```

```python
# Reto extra: Lee tres notas y muestra promedio, mínima, máxima y si aprueba (>=3.0).
print("--- Reto Extra: Analizador de Notas ---")
try:
    nota1 = float(input("Introduce la nota 1: "))
    nota2 = float(input("Introduce la nota 2: "))
    nota3 = float(input("Introduce la nota 3: "))
    
    notas = [nota1, nota2, nota3]
    
    promedio_notas = sum(notas) / len(notas)
    nota_minima = min(notas)
    nota_maxima = max(notas)
    
    # Operador ternario para decidir el estado
    estado = "Aprobado" if promedio_notas >= 3.0 else "Reprobado"
    
    print("\n--- Resumen de Calificaciones ---")
    print(f"Notas: {notas}")
    print(f"Promedio: {promedio_notas:.2f}")
    print(f"Nota mínima: {nota_minima:.2f}")
    print(f"Nota máxima: {nota_maxima:.2f}")
    print(f"Estado: {estado}")
    
except ValueError:
    print("Asegúrate de introducir solo números para las notas.")
print("-" * 20)
```
